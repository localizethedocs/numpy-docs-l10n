A :class:`dtype` object is equal to all data type specifications that are
equivalent to it.


```python
import numpy as np
```

```python
a = np.array([1, 2], dtype=np.float64)
a.dtype == np.dtype(np.float64)
```

```python
a.dtype == np.float64
```

```python
a.dtype == float
```

```python
a.dtype == "float64"
```

```python
a.dtype == "d"
```
