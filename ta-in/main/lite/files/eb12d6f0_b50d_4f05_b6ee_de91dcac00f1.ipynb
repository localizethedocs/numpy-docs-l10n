```python
# Example of providing the optional output parameter
out1 = np.array([0], dtype=np.float64)
out2 = np.sinh([0.1], out1)
out2 is out1
```

```python
# Example of ValueError due to provision of shape mis-matched `out`
np.sinh(np.zeros((3,3)),np.zeros((2,2)))
```
